```python
import scipy.stats as sps
import numpy as np
import ipywidgets as widgets
import matplotlib.pyplot as plt

```

```python
grid = np.linspace(-2, 2, 1000)
```

# Равномерное распределение

```python
plt.plot(grid, sps.uniform.pdf(grid), color='black', 
         lw=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плотность случайной величины')
plt.grid()
plt.show()
```

```python
plt.plot(grid, sps.uniform.cdf(grid), color='black', 
         lw=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция распределения случайной величины')
plt.grid()
plt.show()
```

# Биномиальное распределение

```python
plt.plot(grid, sps.binom(n=10, p=0.6).pmf(grid), color='black', 
         lw=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плотность случайной величины')
plt.grid()
plt.show()
```

```python
plt.plot(grid, sps.binom(n=10, p=0.6).cdf(grid), color='black', 
         lw=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция распределения случайной величины')
plt.grid()
plt.show()
```

# Распределение Пуассона

```python
plt.plot(grid, sps.poisson(0.5).pmf(grid), color='black', 
         lw=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Плотность случайной величины')
plt.grid()
plt.show()
```

```python
plt.plot(grid, sps.poisson(0.5).cdf(grid), color='black', 
         lw=5)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Функция распределения случайной величины')
plt.grid()
plt.show()
```
