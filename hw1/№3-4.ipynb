```python
import math
import scipy.misc
import matplotlib.pyplot as plt
import numpy as np

def fun(x):
    return math.sin(x)
x = list(np.linspace(-10, 10, 1000))
y = [fun(x) for x in x]
d = [scipy.misc.derivative(fun, i, dx=1e-6) for i in x]

plt.plot(x, y, "g.", label="y = f(x)")
plt.plot(x, d, "r.", label="y = f'(x)")
plt.xlabel('x')
plt.ylabel('y')
plt.grid(visible=True, which="both", axis="both", linestyle="-")
plt.legend()
plt.show()
```

```python
from sympy import *
def f(x, y):
    return x ** 2 - y ** 2

x = np.linspace(-100, 100, 50)
y = np.linspace(-100, 100, 50)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)

fig = plt.figure()
ax = plt.axes(projection='3d')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
ax.plot_surface(X, Y, Z)
x = Symbol('x') 
y = Symbol('y') 
expr = x**2 + y**2 
ax.set_title(f'f(x,y)')
plt.show()
```

```python

x = np.linspace(-100, 100, 20)
y = np.linspace(-100, 100, 20)
x1, y1 = np.meshgrid(x, y)
gradX = 2 * x1
gradY = -2 * y1
dirX = gradX / np.sqrt(gradX**2 + gradY**2)
dirY = gradY/ np.sqrt(gradX**2 + gradY**2)

plt.quiver(x1, y1, dirX, dirY, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('  f(x,y)')
plt.show()
```
